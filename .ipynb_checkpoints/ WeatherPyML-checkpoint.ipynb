```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
```

```python
from api_keys import api_key
```

```python
from citipy import citipy
```

```python
output_data_file = "output_data/cities.csv"
```

```python
lat_range = (-90, 90)
lng_range = (-180, 180)
lat_lngs = []
cities = []
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    if city not in cities:
        cities.append(city)
#len(cities)
list_citites_df=pd.DataFrame(cities)
list_citites_df.columns=["City"]
list_citites_df.head()
```

```python
# creating th dataframe to store the data
column = []
cities_df = pd.DataFrame(column)
api = "http://api.openweathermap.org/data/2.5/weather?"
count_of_rows = 1

print(" Beginning Data Retrieval")
for index, row in list_citites_df.iterrows():
    look_city = row["City"]
    look_url = api + "appid=" + api_key + "&units=IMPERIAL" + "&q=" + look_city.replace(" ","+")
    city_data = requests.get(look_url).json()
    if city_data["cod"] == "404":
        print("City not found. Skipping...")

    else:
        cities_df.set_value(index, "City", city_data["name"])
        cities_df.set_value(index, "Country", city_data["sys"]["country"])
        cities_df.set_value(index, "Temperature (F)", city_data["main"]["temp"])
        cities_df.set_value(index, "Latitude", city_data["coord"]["lat"])
        cities_df.set_value(index, "Longitude", city_data["coord"]["lon"])
        cities_df.set_value(index, "Humidity (%)", city_data["main"]["humidity"])
        cities_df.set_value(index, "Cloudiness (%)", city_data["clouds"]["all"])
        cities_df.set_value(index, "Wind Speed (mph)", city_data["wind"]["speed"])
        
        print("--------------------------------------------------")
        print("Proceesing Record: City # " , count_of_rows, ' | ' , city_data["name"], city_data["sys"]["country"])
        count_of_rows += 1
print("--------------------------------------------------")
print("Data Retrieval Complete")
print("--------------------------------------------------")
```

```python
#Plotting!!!!!!  Latitude vs. Temperature Plot
plt.scatter(cities_df["Latitude"], 
            cities_df["Temperature (F)"], c=cities_df["Temperature (F)"],
            edgecolor="black", linewidths=0.5, marker="o", 
            cmap='summer', alpha=0.9, label="City")

plt.title(f"City Latitude vs. Max Temperature {time.strftime('%m/%d/%Y')}")
plt.ylabel("Max Temperature (F)")
plt.xlabel("Latitude")
plt.grid(True)
plt.xlim([-80, 100])
plt.ylim([-60, 120])
          

plt.savefig("WeatherpyFiles/Latitude_vs_Temperature.png")

plt.show()
```
